```python
import pandas as pd
```

```python
df = pd.DataFrame(index=['A', 'B', 'C', 'D', 'E', ])

df['col1'] = [1,2,3,4,0]
df['col2'] = [10,20,30,40,25]
df
```

```python
df = df.apply(lambda x: x*3)
df
```

```python
# UPDATE A SPECIFIC ROW USING AXIS =1

df = df.apply(lambda x: x+5 if x.name in ['B'] else x, axis = 1)
df
```

```python
# UPDATE A SPECIFIC COLUMN USING AXIS =0

df = df.apply(lambda x: x*26 if x.name in['col1'] else x, axis = 0)
df
```

```python
values = [[60,78,65,52,85],
          [75,85,62,61,60],
          [89,76,56,51,35],
          [40,60,70,40,45],
          [50,70,75,46,85],
          [45,66,77,49,78]]

results = pd.DataFrame(values,columns=['English', 'Physics', 'Chemistry', 'Maths', 'Biology'],
                   index = ['Majid', 'Mohsin', 'Suresh', 'Suman', 'Monish', 'Vaani'])
results
```

```python
# ADD 10 GRACE MARKS TO THE MATHS COLUMN

results['Maths'] = results['Maths'].apply(lambda x:x+10)
results
```

```python
results2 = results.apply(lambda x:x*2 if x.name in ['Maths'] else x, axis = 0)
results2
```

```python
# ADD 5 GRACE MARKS TO ALL COLUMNS

results = results.apply(lambda x:x+5)
results
```

```python
# GET A NEW COLUMN CALLED TOTAL MARKS

results['Total'] = results.apply(lambda x: x['English'] + x['Physics'] + x['Chemistry']
 + x['Maths'] + x['Biology'], axis = 1)
results
```

```python
# GET A NEW COLUMN CALLED PERCENTAGE

results['Percentage'] = results.apply(lambda x: x['English'] + x['Physics'] + x['Chemistry']
 + x['Maths'] + x['Biology'], axis = 1)/5
results
```

```python
results.drop(['Total', 'Percentage'], inplace=True, axis=1)
results
```

```python
# MAJID DID NOT OBSERVE INTEGRITY, DEDUCT 5 MARKS FOM EACH SUBJECT

results = results.apply(lambda x:x-5 if x.name in ['Majid'] else x, axis = 1)
results
```

```python
# APPLY LAMBDA BASED ON CONDITIONS

results['Physics'] = results['Physics'].apply(lambda x: x-7 if x>=75 else x)
results
```
