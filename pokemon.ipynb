# MAKING CHANGES TO THE DATA

#### Add a column

```python
poke['Total'] = poke['HP'] + poke['Attack'] + poke['Defense'] + poke['Sp. Atk'] + poke['Sp. Def'] + poke['Speed']

poke.head(5)
```

```python
poke['Total'] = poke.iloc[:, 4:10].sum(axis = 1)

poke.head()
```

```python
cols = list(poke.columns.values)
poke_with_Total = poke[cols[0:4] + [cols[-1]] + cols[4:12]]

poke_with_Total.tail(7)
```

#### SAVING NEW DATAFRAME (DESIRED FORMAT)
